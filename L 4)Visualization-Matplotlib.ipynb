```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y = np.array([10,135,25,15,19])
plt.pie(y)
plt.show()    # prefrence of coloring 1.Blue 2.Orange 3.Green 4.Red 5.Violet
```

```python
### Add labels to the pie chart with the label parameter.

### The label parameter must be an array with one label for each widge
y=np.array([18,25,15,10])
mylabels=["Cherries","Apple","Bananas","Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
### As mentoined the default start angel is at the x-axis,but you can change the start angel
### by specifying a startangle parameter.

### The startangle parameter is defined with an angel in degree, default angle is 0:
y=np.array([25,18,15,10])
mylabels=["Cherries","Apple","Bananas","Dates"]

plt.pie(y, labels = mylabels, startangle = 180)
plt.show()
```

```python
y=np.array([25,18,15,10])
mylabels=["Cherries","Apple","Bananas","Dates"]
myexplode=[0,0.95,0.1,3.2]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
y=np.array([25,18,15,10])
mylabels=["Cherries","Apple","Bananas","Dates"]
mycolors=["black","hotpink","b","yellow"]

plt.pie(y, labels = mylabels, colors=mycolors)
plt.show()
```
