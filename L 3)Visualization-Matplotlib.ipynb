```python
### You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,marker="*",ms=20,mec="r")
plt.show()
```

```python
### use both mec and mfc arguments to color of the entire marker
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,marker="o",ms=20,mec="b",mfc="r")
plt.show()
```

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,linestyle="dashed",marker="o",ms=20,mec="hotpink",mfc="blue")
plt.show()
```

```python
### Set line properties for the grid
### You can also set the line properties for the grid,like this:grid
                          ### (color="color",linestyle="linestyle",linewidth=number)
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.title("Sports Watch Data",loc="left",fontsize=20,color="green")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.plot(x,y)
plt.grid(color="green",linestyle="--",linewidth=2.5)
plt.show()
```

```python
### The scatter() function plots one dot for each observation.it needs two array of the same length
### one for the value of the x-axis,and one for the value of the y-axis.
x=np.array([5,7,8,7])
y=np.array([23,12,45,23])
plt.scatter(x,y)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])
colors=(["red","yellow","black","blue","green"])
size=([10,200,30,40,50])
plt.scatter(x,y,c=colors,s=size)
plt.show()
```
