#### 1. Load the Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
loan = pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/Yhills_July12_Analytics/refs/heads/main/loan_prediction.csv")
```

#### 2. Understand the Data

```python
loan.head()
```

```python
loan.shape
```

```python
loan.isnull().sum()
```

```python
loan.dtypes
```

#### 3. Data Cleaning
##### Data Cleaning = Removing/filling nulls, fixing datatypes, handling duplicates, encoding categories, and sometimes scaling data.

##### convert categorical data to float:
###### We'll use Label Encoding or One-Hot Encoding, and then convert the results to float type.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Label encode categorical columns
le = LabelEncoder()

for col in loan.columns:
    if loan[col].dtype == 'object':
        loan[col] = le.fit_transform(loan[col])

# Convert entire DataFrame to float
loan = loan.astype(float)
```

```python
loan.dtypes
```

##### Handling Missing Values

```python
loan.isnull().sum()
```

```python
loan.head()
```

```python
median1 =loan["LoanAmount"].median()
loan["LoanAmount"].replace(np.nan, median1, inplace=True)
```

```python
median2 =loan["Loan_Amount_Term"].median()
loan["Loan_Amount_Term"].replace(np.nan, median2, inplace=True)
```

```python
median3 =loan["Credit_History"].median()
loan["Credit_History"].replace(np.nan, median3, inplace=True)
```

```python
loan.isnull().sum()
```

#### 4. Checking The Best Fit Model

```python
x = loan.drop("Loan_Status", axis=1)
y = loan[["Loan_Status"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=1)
```

##### Logistic Regression

```python
reg_model = LogisticRegression(solver="liblinear")
reg_model.fit(x_train, y_train)
```

```python
print(reg_model.score(x_train, y_train))
print(reg_model.score( x_test, y_test))
```

##### Decision Tree

```python
dtree=DecisionTreeClassifier(criterion='gini')
dtree.fit(x_train, y_train)
```

```python
print(dtree.score(x_train, y_train))
print(dtree.score( x_test, y_test))
```

```python
# trying to fit the model
dTree=DecisionTreeClassifier(criterion='gini', max_depth=1, random_state=0)
dTree.fit(x_train, y_train)
print(dTree.score(x_train, y_train))
```

```python
y_predict = dTree.predict(x_test)
```

```python
print(dTree.score(x_test, y_test))
```

##### Metrics

```python
predictions=reg_model.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

```python
cm = metrics.confusion_matrix(y_test, predictions, labels=[0,1])

df_cm = pd.DataFrame(cm, index = [i for i in ["NO","YES"]],
                   columns = [i for i in ["NO","YES"]])
plt.close('all')  # ðŸ”„ Clear previous plots
plt.figure(figsize=(7,5))
(sns.heatmap(df_cm, annot=True, fmt="g"))
plt.show()
```

##### AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

# Use decision tree as base learner
base_tree = DecisionTreeClassifier(max_depth=1, random_state=0)

abcl = AdaBoostClassifier(estimator=base_tree, n_estimators=100, learning_rate=0.8, random_state=0)
abcl.fit(x_train, y_train)

print("Train Accuracy:", abcl.score(x_train, y_train))
print("Test Accuracy:", abcl.score(x_test, y_test))
```

##### GradientBoos

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbcl=GradientBoostingClassifier(n_estimators=200, random_state=0)
gbcl.fit(x_train, y_train)
print(gbcl.score(x_train, y_train))
#y_predict=abcl.predict(x_test)
print(gbcl.score( x_test, y_test))
```

##### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfcl=RandomForestClassifier(n_estimators=160, random_state=0, max_features=3)
rfcl.fit(x_train, y_train)
print(rfcl.score(x_train, y_train))
#y_predict=abcl.predict(x_test)
print(rfcl.score( x_test, y_test))
```

```python
models = {
    "Logistic Regression": reg_model.score(x_test, y_test),
    "Decision Tree": dtree.score(x_test, y_test),
    "AdaBoost": abcl.score(x_test, y_test),
    "Gradient Boost": gbcl.score(x_test, y_test),
    "Random Forest": rfcl.score(x_test, y_test)
}

for name, score in models.items():
    print(f"{name}: {score:.4f}")
```

##### -> Result Summary:

###### Model	              Train Accuracy	      Test Accuracy
###### Logistic Regression	      82%	                78%
###### Decision Tree	          81%	                78%
###### AdaBoost	                  88%	                77%
###### Gradient Boost	          98%	                77%
###### Random Forest	         100%	                77%

#### Conclusion:
###### AdaBoost was selected as the final model. It provided the best balance between performance and generalization,
###### avoiding the overfitting issues observed in other ensemble models. 
###### The model is suitable for production deployment and can accurately guide loan approval decisions.
